## 2.1 פונקצייה המחשבת את ההפרש של כל אסון בימים

River Basin - להשאיר רלוונטי להצפות
להוסיף  פונקצייה של זמן בעמודת חישוב שתחשב את ההפרש בין התאריכים שקרו ההצפות כדי לדעת את ממשך הזמן

```python
## הוספת פונקצייה שתחשב את ההפרש שקרו ההצפות כדי לדעת את משך הימים של כל הצפה
import pandas as pd
import numpy as np

# פונקציה לחישוב משך האסון בימים
def calculate_disaster_duration(row):

    try:
        # אם חסר אחד מהערכים → מחזיר NaN
        required_cols = [
            'Start Year',
            'Start Month',
            'Start Day',
            'End Year',
            'End Month',
            'End Day'
        ]

        if row[required_cols].isnull().any():
            return np.nan

        # יצירת תאריך התחלה
        start_date = pd.Timestamp(
            year=int(row['Start Year']),
            month=int(row['Start Month']),
            day=int(row['Start Day'])
        )

        # יצירת תאריך סיום
        end_date = pd.Timestamp(
            year=int(row['End Year']),
            month=int(row['End Month']),
            day=int(row['End Day'])
        )

        # חישוב הפרש ימים
        return (end_date - start_date).days

    except:
        return np.nan


# יצירת עמודה חדשה
df['Disaster Duration (Days)'] = df.apply(
    calculate_disaster_duration,
    axis=1
)
```

## 2.7 יצירת מדד חומרת ועומצת האסון - נירמול לפי הכמות הרוגים,פצועים והנזק הכלכלי לאחר אינפלציה
חלוקת ל-5 קבוצות שונות עם דרכות חומרה שונות

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

severity_cols = [
    'Total Deaths',
    'Total Affected',
    'Adjusted Total Damage'
]

# יצירת עמודות ריקות בדאטה המקורי
df['Severity Score Raw'] = pd.NA
df['Severity Level'] = pd.NA
df['Severity Description'] = pd.NA

# עבודה רק על שורות שיש בהן את כל המדדים
valid_idx = df.dropna(subset=severity_cols).index

scaler = MinMaxScaler()

scaled_values = scaler.fit_transform(
    df.loc[valid_idx, severity_cols]
)

# ציון חומרה רציף
df.loc[valid_idx, 'Severity Score Raw'] = scaled_values.mean(axis=1)

# דירוג 1–5
df.loc[valid_idx, 'Severity Level'] = pd.qcut(
    df.loc[valid_idx, 'Severity Score Raw'],
    q=5,
    labels=[1, 2, 3, 4, 5]
)

severity_labels = {
    1: 'Mild',
    2: 'Moderate',
    3: 'Severe',
    4: 'Very Severe',
    5: 'Catastrophic'
}

df.loc[valid_idx, 'Severity Description'] = (
    df.loc[valid_idx, 'Severity Level']
    .astype(int)
    .map(severity_labels)
)
```

2.7 הוספת מדד הצפיפות בכל מדינה נלקח מאתר קבוצת הבנק העולמי
קישור: https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd

# קריאת קובץ האוכלוסייה
population = pd.read_csv("popultion disasters.csv")

# ניקוי רווחים משמות עמודות
population.columns = population.columns.astype(str).str.strip()

# זיהוי אוטומטי של עמודות השנים
year_cols = [
    col for col in population.columns
    if col.isdigit()
]

# הפיכת השנים לעמודה אחת
population_long = population.melt(
    id_vars=['Country Name', 'Country Code'],
    value_vars=year_cols,
    var_name='Year',
    value_name='Population'
)

# שינוי שמות כדי להתאים לדאטה שלך
population_long = population_long.rename(columns={
    'Country Name': 'Country',
    'Country Code': 'Country Code'
})

# המרות סוגי נתונים
population_long['Year'] = pd.to_numeric(population_long['Year'], errors='coerce')
population_long['Population'] = pd.to_numeric(population_long['Population'], errors='coerce')

# התאמת שמות מדינות בדאטה הראשי
df['Country'] = df['Country'].replace({
    'United States of America (the)': 'United States',
    'Viet Nam': 'Vietnam',
    'Philippines (the)': 'Philippines'
})

# מיזוג לפי מדינה ושנה
df = df.merge(
    population_long[['Country', 'Year', 'Population']],
    on=['Country', 'Year'],
    how='left'
)

# יצירת GDP לנפש
df['GDP per Capita'] = df['GDP'] / df['Population']

# בדיקה
df[['Country', 'Year', 'GDP', 'Population', 'GDP per Capita']].head(20)
```
